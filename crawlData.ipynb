# Import các thư viện

```python
from selenium import webdriver
from requests_html import HTML
import time
```

# Mô tả file data.csv

#### File data.csv có các trường cách nhau bằng tab `\t`


- `id`: ID của khóa học
-`title`: Tên của khóa học
-`provider`: Tên nơi cung cấp khóa học, nếu có nhiều hơn một nơi thì sẽ ngăn cách nhau bằng dấu `,`
-`score`: Rating của khóa học
-`rating`: Số lượng người rating
-`level`: Level của khóa học, thông thường thì sẽ có `level` cho mỗi trang web của khóa học, nếu không có trong trang web của khóa học thì coursera đã quy định `level = mixed`
-`duration`: Thời gian để hoàn thành khóa học
-`language`: Ngôn ngữ mà khóa học sử dụng
-`skills`: Những kỹ năng mà khóa học hướng đến
-**`enroll`**: Số lượng người tham gia khóa học. Nếu không có số lượng người tham gia khóa học đó thì ta sẽ không crawl

# Khởi tạo browser và crawl dữ liệu

```python
output_file = open("data.csv", mode='w', encoding='utf-8')
output_file.write("id\ttitle\tprovider\tscore\trating\tlevel\tduration\tlanguage\tskills\tenroll\n")
input_file = open("course_link.txt", "r", encoding="utf-8")

browser = webdriver.Chrome(executable_path='chromedriver.exe')
for index, line in enumerate(input_file):
    if "/learn/" not in line:
        continue
    
    # Go to link
    print(line.strip())
    browser.get(line.strip())
    #browser.get("https://www.coursera.org/learn/leadership-marketing")
    
    # Parse HTML
    r = HTML(html=browser.page_source)
    
    
    # Check number of enroll
    enroll = r.find("div.rc-ProductMetrics strong span")
    if enroll == []:
        continue
    enroll = enroll[0].text
    
    
    # Course Name
    e = r.find(".Banner")[0]
    name = e.find("h1")[0].text

    # University
    schoolBox = e.find(".partnerBanner_np2ice-o_O-Box_120drhm-o_O-displayflex_poyjc")[0]
    schoolNames = []
    if schoolBox.find("img") != []:
        #output_file.write( + '\t')
        schoolNames = schoolBox.find("img")[0].attrs["alt"]
    else:
        schoolList = schoolBox.find("span")
        for school in schoolList:
            #output_file.write(school.text + ',')
            schoolNames.append(school.text)
        schoolNames = ','.join(schoolNames)
        #output_file.write('\t')

    # Score and Rating
    spanlist = e.find("li span")
    score = 0
    rating = 0
    if len(spanlist) != 0:
        #output_file.write(spanlist[0].text + '\t')
        #output_file.write(spanlist[1].text.replace(" ratings",""))
        score = spanlist[0].text
        rating = spanlist[1].text.replace(" ratings","")
    #else:
        #output_file.write('0\t')
        #output_file.write('0\t')
    
    product = r.find("div.ProductGlance")[0]
    product = product.find("div.Box_120drhm-o_O-displayflex_poyjc")
    count = len(product)
    # Level
    level = product[count - 3].find("h4")[0].text
    if "Level" not in level:
        level = "Mixed"
    
    # Duration
    duration = product[count - 2].find("h4")[0].text
    
    # Language
    language = product[count - 1].find("h4")[0].text
    
    # Skills
    skillBox = r.find("div.Skills")
    skillList = []
    if skillBox != []:
        skillSpan = skillBox[0].find("span")
        for i in skillSpan:
            skillList.append(i.text)
    skillList = ','.join(set(skillList))
    
#     print("enroll = ", enroll)
#     print("name = ", name)
#     print("school = ", schoolNames)
#     print("score = ", score)
#     print("rating = ", rating)
#     print("level = ", level)
#     print("duration = ", duration)
#     print("language = ", language)
#     print("skill = ", skillList)
    output_file.write(f"{index+1}\t\
                      {name}\t\
                      {schoolNames}\t\
                      {score}\t\
                      {rating}\t\
                      {level}\t\
                      {duration}\t\
                      {language}\t\
                      {skillList}\t\
                      {enroll}\t\n")
output_file.close()
input_file.close()
```
